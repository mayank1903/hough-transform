```python
import matplotlib.pyplot as plt
import numpy as np

from hough_transform.src.constants import img_path, document_path
from hough_transform.src.utils import plot_image_with_lines_p_hough, plot_image_with_lines_hough, generalPipeline
```

## Standard Hough Transform

```python
from skimage.io import imread
from skimage.feature import canny
from skimage.transform import hough_line, hough_line_peaks
```

```python
# Load the image
image_path = img_path  # Replace with the actual path to your image
image = imread(image_path, as_gray=True)

# Apply edge detection (Canny)
edges = canny(image, sigma=2, low_threshold=0.1, high_threshold=0.3)

# Apply Standard Hough Transform
h, theta, d = hough_line(edges)

# Find the most prominent lines
lines = hough_line_peaks(h, theta, d)

# Plot the original image and the image with detected lines
plot_image_with_lines_hough(image, lines)
```

## Probabilistic Hough Transform

```python
from skimage.transform import probabilistic_hough_line

# Load the image
image_path = img_path  # Replace with the actual path to your image
image = imread(image_path, as_gray=True)

# Apply edge detection (Canny)
edges = canny(image, sigma=2, low_threshold=0.1, high_threshold=0.3)

# Apply Probabilistic Hough Transform
min_line_length = 75
max_line_gap = 5
lines = probabilistic_hough_line(edges, threshold=10, line_length=min_line_length, line_gap=max_line_gap)

# Plot the original image and the image with detected lines
plot_image_with_lines_p_hough(image, lines)
```
